Step 5. Create a pie chart presenting the male/female proportion

```python
# Đọc dữ liệu từ file CSV
titanic = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv'
data = pd.read_csv(titanic)

# Đặt chỉ mục là PassengerId
data.set_index('PassengerId', inplace=True)

# Tính tỉ lệ giữa số lượng nam và nữ
sex_counts = data['Sex'].value_counts()
male_proportion = sex_counts['male'] / sex_counts.sum()
female_proportion = sex_counts['female'] / sex_counts.sum()

# Vẽ biểu đồ tròn
labels = ['Male', 'Female']
sizes = [male_proportion, female_proportion]
colors = ['blue', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
# lọc dữ liệu theo giới tính
male_data = data[data['Sex'] == 'male']
female_data = data[data['Sex'] == 'female']

# vẽ biểu đồ
fig, ax = plt.subplots()
ax.scatter(male_data['Age'], male_data['Fare'], c='blue', label='Male')
ax.scatter(female_data['Age'], female_data['Fare'], c='red', label='Female')
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_title('Fare paid vs Age by Gender')
ax.legend()
plt.show()
```

Step 7. How many people survived

```python
survived_count = len(data[data['Survived'] == 1])
print("The number of people who survived: ", survived_count)
```

Step 8. Create a histogram with the Fare payed

```python
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv'
data = pd.read_csv(titanic)

# Create the histogram
plt.hist(data['Fare'], bins=20)

# Add labels and title
plt.xlabel('Fare paid')
plt.ylabel('Frequency')
plt.title('Histogram of Fare paid')

# Display the histogram
plt.show()
```

## bonus:  tỷ lệ sống sót giữa nam và nữ trên tàu Titanic 

```python
# Tính tỷ lệ sống sót
survived_by_sex = data.groupby('Sex')['Survived'].mean()

# Vẽ biểu đồ cột
plt.bar(survived_by_sex.index, survived_by_sex.values)

# Đặt tên cho các trục
plt.xlabel('Giới tính')
plt.ylabel('Tỷ lệ sống sót')

# Đặt tiêu đề cho biểu đồ
plt.title('Tỷ lệ sống sót giữa nam và nữ trên tàu Titanic')

# Hiển thị biểu đồ
plt.show()
```
