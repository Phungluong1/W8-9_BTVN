Step 4. See the first 10 entries

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ URL và lưu vào DataFrame
df = pd.read_csv(chipo, sep='\t')

# Hiển thị 5 dòng đầu tiên của DataFrame
print(df.head(10))
```

 Step 5. Create a histogram of the top 5 items bought

```python
# Lấy 5 sản phẩm đầu tiên
top_5_items = df['item_name'].head(5)
print(top_5_items)
# Lọc ra các dòng chứa 5 sản phẩm này
df_top_5_items = df[df['item_name'].isin(top_5_items)]
# Tạo histogram cho số lượng sản phẩm được đặt hàng
plt.hist(df_top_5_items['item_name'], bins=5)
plt.title('Số lượng sản phẩm được đặt hàng (5 sản phẩm đầu tiên)')
plt.xlabel('Sản phẩm')
plt.ylabel('Số lượng')
plt.show()

# # lấy 5  sản phẩm được bán nhiều nhất
# top_5_items = df['item_name'].value_counts().head(5).index
# # Lọc ra các dòng chứa 5 sản phẩm này
# df_top_5_items = df[df['item_name'].isin(top_5_items)]
# # Tạo histogram cho số lượng sản phẩm được đặt hàng
# plt.hist(df_top_5_items['item_name'], bins=5)
# plt.title('Số lượng sản phẩm được đặt hàng (5 sản phẩm bán nhiều nhất)')
# plt.xlabel('Sản phẩm')
# plt.ylabel('Số lượng')
# plt.show()
```

Step 6. Create a scatterplot with the number of items orderered per order price

```python
import pandas as pd
import matplotlib.pyplot as plt
# Đọc dữ liệu từ chipo và lưu vào DataFrame
df = pd.read_csv(chipo, sep='\t')

item_price = []
quantity = []
# lấy ra đơn giá
for i in range(4622):
  item_price.append(df['item_price'][i][1:])
# lấy ra số lượng
for i in range(4622):
  quantity.append(df['quantity'][i])
# ép kiểu object về float
quantity=list(map(float, quantity))
item_price=list(map(float, item_price))
print('đơn giá')
print(item_price)
print('số lượng')
print(quantity)
# tính giá trị của đơn hàng dựa trên số lượng và đơn giá
order_price = []
for i in range(4622):
  order_price.append(quantity[i]*item_price[i])
print('giá trị đơn hàng')
print(order_price)
# # Vẽ đồ thị scatterplot
plt.scatter(x=order_price, y=quantity)
plt.title('Số lượng sản phẩm được đặt hàng theo giá trị đơn hàng')
plt.xlabel('Giá trị đơn hàng')
plt.ylabel('Số lượng sản phẩm')
plt.show()
```

```python
plt.scatter(x=item_price, y=quantity)
plt.title('Số lượng sản phẩm được đặt hàng theo đơn giá')
plt.xlabel('Đơn giá mặt hàng')
plt.ylabel('Số lượng sản phẩm')
plt.show()
```
