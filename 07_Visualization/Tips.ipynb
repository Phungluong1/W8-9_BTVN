Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

# Step 2. Import the dataset from this address.
## Step 3. Assign it to a variable called tips

```python
tips = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv'
```

Step 4. Delete the Unnamed 0 column

```python
data = pd.read_csv(tips)
data = data.drop(columns=['Unnamed: 0'])
data
```

Step 5. Plot the total_bill column histogram

```python
data['total_bill'].hist()
plt.show()
```

Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(data['total_bill'], data['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relationship between Total Bill and Tip')
plt.show()
```

Step 7. Create one image with the relationship of total_bill, tip and size.

```python
plt.scatter(data['total_bill'], data['tip'],s=data['size']*10,c=data['size'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relationship between Total Bill, Tip and size')
plt.show()
```

Step 8. Present the relationship between days and total_bill value

```python
plt.scatter(data['day'],data['total_bill'])
plt.xlabel('days')
plt.ylabel('total_bill')
plt.title('present the relationship between days and total_bill value')
```

Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
data = pd.read_csv(tips)
data['sex'].replace('Female', 1, inplace=True)
data['sex'].replace('Male', 3, inplace=True)

plt.scatter(data['tip'],data['day'],c=data['sex'],s=data['sex']*50)
plt.xlabel('days')
plt.ylabel('total_bill')
plt.title('present the relationship between days and total_bill value')
```

Step 10. Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
# Táº¡o biá»ƒu Ä‘á»“ boxplot
fig, ax = plt.subplots(figsize=(8, 6))
data.boxplot(column='total_bill', by=['day', 'time'], ax=ax)
ax.set_title('Total Bill by Day and Time')
ax.set_xlabel('(Day, Time)')
ax.set_ylabel('Total Bill')
plt.show()
```

Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tip dataset
tip = sns.load_dataset('tips')

# Create a FacetGrid object
g = sns.FacetGrid(tip, col='time')

# Map the histograms of tip to the FacetGrid object
g.map(plt.hist, 'tip')

# Display the plot
plt.show()
```

Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
# Filter data for male and female separately
male_data = data[data['sex'] == 'Male']
female_data = data[data['sex'] == 'Female']

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Plot scatterplot for male customers
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=male_data, ax=ax1)
ax1.set_title('Male Customers')

# Plot scatterplot for female customers
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=female_data, ax=ax2)
ax2.set_title('Female Customers')

# Display the plot
plt.show()
```
