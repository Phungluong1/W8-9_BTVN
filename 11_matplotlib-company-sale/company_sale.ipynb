Exercise 1: Read Total profit of all months and show it using a line plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
data = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
data = pd.read_csv(data)

# Hiển thị tổng lợi nhuận bằng biểu đồ dòng
months = data['month_number']
plt.plot(months, data['total_profit'])
plt.xlabel('month number')
plt.ylabel('profit')
plt.title('company profit per month')
plt.show()
```

Exercise 2: Get total profit of all months and show line plot with the following Style properties

```python
import pandas as pd
import matplotlib.pyplot as plt

data = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(data)

monthly_profit = df.groupby('month_number')['total_profit'].sum()
plt.plot(monthly_profit.index, monthly_profit.values, color='red', linestyle='--', marker='o', linewidth=2,markerfacecolor='black',label = 'profit data of last year')

plt.title('Company sales data of last year')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(range(1, 13))
plt.yticks([100000,200000,300000,400000,500000])
plt.legend(loc='lower right', fontsize=12, title_fontsize=14)

plt.show()
```

Exercise 3: Read all product sales data and show it using a multiline plot

```python
import pandas as pd
import matplotlib.pyplot as plt
# đọc dữ liệu
data = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(data)

# vẽ biểu đồ đường cho face cream
monthly_profit = df.groupby('month_number')['facecream'].sum()
plt.plot(monthly_profit.index, monthly_profit.values, color='blue', linestyle='-', marker='o', linewidth=2,label = 'facecream')
# vẽ biểu đồ đường cho face wash
monthly_profit = df.groupby('month_number')['facewash'].sum()
plt.plot(monthly_profit.index, monthly_profit.values, color='yellow', linestyle='-', marker='o', linewidth=2,label = 'facewash')
# vẽ biểu đồ đường cho tooth paste
monthly_profit = df.groupby('month_number')['toothpaste'].sum()
plt.plot(monthly_profit.index, monthly_profit.values, color='green', linestyle='-', marker='o', linewidth=2,label = 'toothpaste')
#vẽ biểu đồ đường cho bathing soap
monthly_profit = df.groupby('month_number')['bathingsoap'].sum()
plt.plot(monthly_profit.index, monthly_profit.values, color='red', linestyle='-', marker='o', linewidth=2,label = 'bathingsoap')
# vẽ biểu đồ đường cho shampoo
monthly_profit = df.groupby('month_number')['shampoo'].sum()
plt.plot(monthly_profit.index, monthly_profit.values, color='purple', linestyle='-', marker='o', linewidth=2,label = 'shampoo')
# vẽ biểu đồ đường cho moisturizer
monthly_profit = df.groupby('month_number')['moisturizer'].sum()
plt.plot(monthly_profit.index, monthly_profit.values, color='brown', linestyle='-', marker='o', linewidth=2,label = 'moisturizer')

plt.title('Company sales data of last year')
plt.xlabel('Month Number')
plt.ylabel('Sale until in number')
plt.xticks(range(1, 13))
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.legend(loc='upper left', fontsize=12, title_fontsize=11)
# biểu đồ moisturizer đè lên facewash do dữ liệu trùng nhau
plt.show()
```

Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data
data = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(data)

# Extract toothpaste sales data of each month
toothpaste_data = df[['month_number', 'toothpaste']]

# Create scatter plot
plt.scatter(toothpaste_data['month_number'], toothpaste_data['toothpaste'],label = 'tooth paste sales data')
plt.title('Tooth paste sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of untils sold')
plt.legend(loc='upper left', fontsize=12, title_fontsize=11)
plt.grid(color = 'black', linestyle = '--', linewidth = 1)
plt.show()
```

Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(data)

# Select face cream and facewash sales data
fc_sales = df['facecream']
fw_sales = df['facewash']

# Create a bar chart
plt.bar(df['month_number']-0.3, fc_sales, width=0.3,align='edge',label='Face Cream sales data')
plt.bar(df['month_number'], fw_sales, width=0.3,align='edge',label='Face Wash sales data')

# Set chart title and axis labels
plt.title('Face Cream and Face Wash sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales until in number')
plt.xticks(range(1, 13))
plt.grid(color = 'black', linestyle = '--', linewidth = 1)

# Set legend
plt.legend(loc='upper left')

# Show the chart
plt.show()
```

Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
import pandas as pd
import matplotlib.pyplot as plt

# read the data
data = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(data)

# group data by month and get total profit
monthly_profit = df.groupby('month_number')['total_profit'].sum()

# create histogram of profit data
plt.hist(monthly_profit, bins=10, color='blue',label='Profit data')

# set labels and title
plt.xlabel('Profit range in data')
plt.ylabel('Actual Profit in dollar')
plt.title('Profit data')
plt.legend(loc='upper left')

# display the plot
plt.show()
```

Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')

# Extract the data for Bathing Soap and Facewash
bathing_soap_data = data.groupby('month_number')['bathingsoap'].sum()
facewash_data = data.groupby('month_number')['facewash'].sum()

# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 4))

# Plot the data for Bathing Soap on the first subplot
ax1.plot(bathing_soap_data.index, bathing_soap_data.values, color='black',marker='o')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales')
ax1.set_yticks([7500,10000,12500])
ax1.set_title('Bathing Soap Sales')

# Plot the data for Facewash on the second subplot
ax2.plot(facewash_data.index, facewash_data.values, color='red',marker='o')
ax2.set_xlabel('Month')
ax2.set_ylabel('Sales')
ax2.set_yticks([1500,2000])
ax2.set_title('Facewash Sales')

# Show the plot
plt.show()
```
